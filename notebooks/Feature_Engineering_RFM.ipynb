```python
import pandas as pd
```

```python
base = pd.read_csv('../data/processed/base_tratada.csv')
base.head()
```

```python
# Convertendo a coluna InvoiceDate para o tipo datetime
base['InvoiceDate'] = pd.to_datetime(base['InvoiceDate'])
```

```python
# Para criar o RFM, precisa de uma data de referência
data_referencia = base['InvoiceDate'].max() 
print(data_referencia)
```

```python
Recency = data_referencia
```

```python
# Criar tabela RFM
rfm = base.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (Recency - x.max()).days,  
    'InvoiceNo': 'nunique',
    'TotalPrice': 'sum'
    })
rfm.columns = ['Recency', 'Frequency', 'Monetary']
```

```python
print(rfm.head())
```

```python
# Ver as informações estatísticas da tabela RFM
print(rfm.describe())
print(rfm.info())
```

```python
# Resetar o índice
rfm.reset_index(inplace=True)
rfm.head()
```

```python
# Salvar a base RFM em um arquivo CSV
rfm.to_csv('../data/processed/rfm_clientes.csv', index=False)
```
